```python
import pandas as pd
from statsmodels.formula.api import mixedlm
import pandas as pd


data = pd.read_csv('data_for_error_analysis.csv')
```

```python
model1 = mixedlm('''correct ~ + unusual_insult + historical_reference + call_for_reconciliation
       + neutral_sentiment + negative_text_without_ethnic_conflict + sarcasm + contrast 
       + migrantophobia + secondary_topic + model_2 + model_3 + has_eth_conflict_raw
      ''', 
      data, groups=data["ID"]).fit(reml=False)

model2 = mixedlm('''correct ~ + (unusual_insult + historical_reference + call_for_reconciliation
       + neutral_sentiment + negative_text_without_ethnic_conflict + sarcasm + contrast 
       + migrantophobia + secondary_topic + model_2 + model_3) * has_eth_conflict_raw
      ''',
      data, groups=data["ID"]).fit(reml=False)

model3 = mixedlm('''correct ~  + (unusual_insult + historical_reference + call_for_reconciliation
       + neutral_sentiment + negative_text_without_ethnic_conflict + sarcasm + contrast 
       + migrantophobia + secondary_topic)*(model_2 + model_3) + has_eth_conflict_raw
      ''',
      data, groups=data["ID"]).fit(reml=False)

model4 = mixedlm('''correct ~ + ((unusual_insult + historical_reference + call_for_reconciliation
       + neutral_sentiment + negative_text_without_ethnic_conflict + sarcasm + contrast 
       + migrantophobia + secondary_topic)*(model_2 + model_3)) * has_eth_conflict_raw
      ''',
      data, groups=data["ID"]).fit(reml=False)
```

```python
print(f"model1 BIC: {model1.bic}")
print(f"model2 BIC: {model2.bic}")
print(f"model3 BIC: {model3.bic}")
print(f"model4 BIC: {model4.bic}")
```

```python
model2.summary()
```

```python
result=model2
var_resid = result.scale
var_random_effect = float(result.cov_re.iloc[0])
var_fixed_effect = result.predict(data).var()

total_var = var_fixed_effect + var_random_effect + var_resid
marginal_r2 = var_fixed_effect / total_var
conditional_r2 = (var_fixed_effect + var_random_effect) / total_var
marginal_r2, conditional_r2
```
